```python
def odds_of_success(level_remain, target_spdfld):
    cumulated_odds = 0
    max_i = target_spdfld // 3
    for i in range(max_i + 1):
        max_j = (target_spdfld - 3*i) // 2
        for j in range(max_j + 1):
            max_k = min((target_spdfld - 3*i - 2*j), level_remain - i - j)
            for k in range(max_k + 1):
                l = level_remain - i - j - k
                arrange = 1
                for r in range(2, level_remain+1):
                    arrange *= r
                for r in range(1, i+1):
                    arrange /= r
                for r in range(1, j+1):
                    arrange /= r
                for r in range(1, k+1):
                    arrange /= r
                for r in range(1, l+1):
                    arrange /= r
                #print(i, j, k, l, arrange)
                single_odds = (arrange * (4**i) * (9**j) * (12**k) * (10**l)) / (35**level_remain)
                #print(single_odds)
                cumulated_odds += single_odds
    return cumulated_odds
```

```python
for j in range(4, 11):
    for i in range(10, -1, -1):
        print(j, i, 100*odds_of_success(j, i))
```
